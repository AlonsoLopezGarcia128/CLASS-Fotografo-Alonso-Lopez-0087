```python
# Sistema de Gestión de Fotógrafos

fotografos = []

def agregar_fotografo():
    id_f = input("Introduce el ID del fotógrafo: ")
    nombre = input("Nombre del fotógrafo: ")
    especialidad = input("Especialidad (Boda, Retrato, Paisaje): ")
    fotografos.append({
        "id": id_f,
        "nombre": nombre,
        "especialidad": especialidad
    })
    print("¡Fotógrafo agregado con éxito!")

def leer_fotografo(f):
    print(f"ID: {f['id']} | Nombre: {f['nombre']} | Especialidad: {f['especialidad']}")

def listar_fotografos():
    if not fotografos:
        print("No hay fotógrafos registrados.")
    else:
        print("\n--- Lista de Fotógrafos ---")
        for f in fotografos:
            leer_fotografo(f)

def buscar_fotografo():
    id_busqueda = input("Introduce el ID del fotógrafo a buscar: ")
    for f in fotografos:
        if f['id'] == id_busqueda:
            leer_fotografo(f)
            return f
    print("Fotógrafo no encontrado.")
    return None

def modificar_fotografo():
    f = buscar_fotografo()
    if f:
        print("Deja en blanco si no deseas cambiar el valor.")
        nuevo_nombre = input(f"Nuevo nombre ({f['nombre']}): ")
        nueva_especialidad = input(f"Nueva especialidad ({f['especialidad']}): ")

        if nuevo_nombre: f['nombre'] = nuevo_nombre
        if nueva_especialidad: f['especialidad'] = nueva_especialidad
        print("Datos actualizados.")

def borrar_fotografo():
    id_borrar = input("Introduce el ID del fotógrafo a eliminar: ")
    global fotografos
    for f in fotografos:
        if f['id'] == id_borrar:
            fotografos.remove(f)
            print("Fotógrafo eliminado correctamente.")
            return
    print("No se encontró el ID.")

# Menú Principal con While
while True:
    print("\n--- SISTEMA DE CONTROL DE FOTÓGRAFOS ---")
    print("1. Agregar fotógrafo")
    print("2. Listar todos")
    print("3. Buscar por ID")
    print("4. Modificar fotógrafo")
    print("5. Borrar fotógrafo")
    print("6. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        agregar_fotografo()
    elif opcion == "2":
        listar_fotografos()
    elif opcion == "3":
        buscar_fotografo()
    elif opcion == "4":
        modificar_fotografo()
    elif opcion == "5":
        borrar_fotografo()
    elif opcion == "6":
        print("Saliendo del sistema...")
        break
    else:
        print("Opción no válida, intenta de nuevo.")
```
